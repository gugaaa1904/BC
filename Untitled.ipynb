### Data Description
Dataset related to customer behavior and preferences in a hotel context. Contains historical booking patterns, revenue data, and customer-specific preferences.

#### Customer Information:
- **ID**: Unique customer identifier (numeric)
- **Nationality**: Customer nationality (categorical: e.g., 'BRA', 'USA')
- **Age**: Customer age (numeric)
- **DaysSinceCreation**: Days since the customer account was created (numeric)
- **NameHash**: Anonymized hash of the customer’s name (text)
- **DocIDHash**: Anonymized hash of the customer’s ID document (text)
- **LodgingRevenue**: Total lodging revenue (numeric)
- **OtherRevenue**: Total revenue from non-lodging services (e.g., food, spa) (numeric)
- **MarketSegment**: Customer market segment (e.g., 'corporate', 'leisure') (categorical)

#### Booking Behaviour:
- **AverageLeadTime**: Average days between booking and arrival date (numeric)
- **BookingsCanceled**: Total canceled bookings by the customer (numeric)
- **BookingsNoShowed**: Total "no-show" bookings (numeric)
- **BookingsCheckedin**: Total bookings that resulted in stays (numeric)
- **PersonNights**: Total person-nights stayed (adults + children × nights) (numeric)
- **RoomNights**: Total room-nights stayed (rooms × nights) (numeric)
- **DistributionChannel**: Booking distribution channel (e.g., 'direct', 'agency') (categorical)

#### Customer Preferences (Binary: 0 = No, 1 = Yes)
- **SRHighFloor**: Requests rooms on higher floors
- **SRLowFloor**: Requests rooms on lower floors
- **SRMediumFloor**: Requests rooms on middle floors
- **SRAccessibleRoom**: Requests accessible rooms
- **SRBathtub**: Requests rooms with bathtubs
- **SRShower**: Requests rooms with showers
- **SRCrib**: Requests cribs
- **SRKingSizeBed**: Requests king-size beds
- **SRTwinBed**: Requests twin beds
- **SRNearElevator**: Requests rooms near elevators
- **SRAwayFromElevator**: Requests rooms away from elevators
- **SRNoAlcoholInMiniBar**: Requests mini-bars without alcohol
- **SRQuietRoom**: Requests quiet rooms

## Data Preparation

Outliers in Age

```python
# Filter the dataset to keep rows with a valid Age between 0 and 100, or missing values.
data = data[data['Age'].isna() | ((data['Age'] >= 0) & (data['Age'] <= 100))]
```

outliers in (PersonsNights < RoomNights)

```python
# Identify rows with errors (PersonsNights < RoomNights)
invalid_rows = data[data['PersonsNights'] < data['RoomNights']]

# Count the number of error rows
n_invalid = invalid_rows.shape[0]

# Delete the error rows from the dataset
data = data[data['PersonsNights'] >= data['RoomNights']]

# Print the number of rows eliminated
print(f"Rows eliminated: {n_invalid}")
```

Eliminate DocIDHash that are duplicated


Basicamente tentei salvar algumas DocIDHash (não sei se está certo, o que é que vcs acham?) Basicamente a primeira cell verifica se há valores duplicados, mas que tenham somente um CheckIn (provavelmente este valor está certo, porque aqui eles foram obrigados a dar os seus dados verdadeiros).
Na cell 2 foi basicamente a fazer o código para guardar os DocId que correspondem a isto. O 3 apaga os DocIDHash que não foram necessários. (não sei se expliquei bem aqui, qualquer cena liguem me).

```python
# Filter rows with duplicated DocIDHash values (i.e. DocIDHash appears in more than one row)
duplicated_docs = data[data.duplicated(subset='DocIDHash', keep=False)]

# Group by DocIDHash and count how many rows in each group have BookingsCheckedIn not equal to 0
doc_checkin_count = duplicated_docs.groupby('DocIDHash')['BookingsCheckedIn'].apply(lambda x: (x != 0).sum())

# Identify groups that exactly match the condition: exactly one row with non-zero BookingsCheckedIn
valid_docs = doc_checkin_count[doc_checkin_count == 1]

# Print a concise message based on the result
if not valid_docs.empty:
    print("Duplicated DocIDHash groups with exactly one non-zero BookingsCheckedIn:")
    print(valid_docs)
else:
    print("No duplicated DocIDHash groups found with exactly one non-zero BookingsCheckedIn.")
```

```python
# Count all occurrences per DocIDHash
all_counts = data['DocIDHash'].value_counts()

# Count occurrences for rows where BookingsCheckedIn is non-zero
nonzero_counts = data[data['BookingsCheckedIn'] != 0]['DocIDHash'].value_counts()

# Valid DocIDHash values are those that appear more than once and have exactly one non-zero BookingsCheckedIn
valid_ids = all_counts[(all_counts > 1) &
                       (all_counts.index.isin(nonzero_counts.index)) &
                       (nonzero_counts == 1)].index

print("Valid duplicated DocIDHash values with exactly one non-zero BookingsCheckedIn:")
#print(list(valid_ids))
```

```python
# Count rows to be removed: rows for valid_ids with BookingsCheckedIn equal to 0
rows_removed = data[(data['DocIDHash'].isin(valid_ids)) & (data['BookingsCheckedIn'] == 0)].shape[0]

# Remove those rows from the dataset (keeping the one row with non-zero BookingsCheckedIn for each valid DocIDHash)
data = data[~((data['DocIDHash'].isin(valid_ids)) & (data['BookingsCheckedIn'] == 0))]

print(f"Rows removed: {rows_removed}")
```

Delete the rest of the duplicated DocIDHash

```python
# Store the Doriginal number of rows
original_rows = data.shape[0]

# Keep only rows where the DocIDHash is unique (i.e. appears exactly once)
data = data[data['DocIDHash'].map(data['DocIDHash'].value_counts()) == 1]

# Calculate and print the number of rows removed
rows_removed = original_rows - data.shape[0]
print(f"Rows removed: {rows_removed}")
```
